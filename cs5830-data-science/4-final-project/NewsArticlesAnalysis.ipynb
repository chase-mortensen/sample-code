```
import glob
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
import re
from nltk.tokenize import word_tokenize
import seaborn as sns
from collections import Counter
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
```

## CNN

```
cnnfiles = []
for file in glob.glob('/content/drive/My Drive/Academics/0002. Spring 2020/CS 6830 Data Science/Final Project/Final Project To Submit/NewsArticles/CNN/*.txt'):
    cnnfiles.append(file)
cnnfiles
```

```
cnnContent = []
cnnContentStemmed = []
for file in cnnfiles:
  
  f = open(file, 'r')
  content = f.read()
  
  cleaned = re.sub(r'[^(a-zA-Z)\s]','', content)
  tokenized = word_tokenize(cleaned)
  stopped = [w for w in tokenized if not w.lower() in stop_words]
  
  stemmedWords = []
  for w in stopped: 
    stemmedWords.append((ps.stem(w)))

  cnnContent.append(stopped)
  cnnContentStemmed.append(stemmedWords)
  f.close()

cnnContent
```

## Counting the most frequent words for articles

```
cnn.text
```

```
from collections import Counter 

orgs = [cnn,nyt,wsj,fox]
orgsLiteral = []
i=0
allorgtext = []
for o in orgs:
  wholetext = []
  for line in o.text:
    wholetext = wholetext + line
  counter = Counter(wholetext)
  most_occured = counter.most_common(100)
  orgsLiteral.append( most_occured)
  i+=1
  allorgtext = allorgtext + wholetext

counter = Counter(allorgtext)
most_occured = counter.most_common(100)
orgsLiteral.append( most_occured)

cnnWordCount=orgsLiteral[0]
nytWordCount=orgsLiteral[1]
wsjWordCount=orgsLiteral[2]
foxWordCount=orgsLiteral[3]
allWordCount=orgsLiteral[4]
```

```
allWordDist = nltk.FreqDist(w.lower() for w in allorgtext)
mostCommon= allWordDist.most_common(100)
mostCommon
```

```
##Removin common words

commonWords=['said','people','states','mr','coronavirus','new','orders','state','trump','governors','gov', 'would','could','according','told','']
  
# removing tuples for condition met 
removingCommon = list(filter(lambda x: x[0] not in commonWords, mostCommon)) 
  
# printing resultant tuple list 
# print ("Resultant tuple list: ", str(removingCommon)) 
removingCommon
```

```
stopwords = nltk.corpus.stopwords.words('english')
allWordExceptStopDist = nltk.FreqDist(w.lower() for w in allorgtext if w not in stopwords)
mostCommon= allWordExceptStopDist.most_common(50)
mostCommon
```

##Creaing list of top subjects for each news sources

```
cnntopWords=[]
for aTuple in cnnWordCount:
    cnntopWords.append(str(aTuple[0]))

nyttopWords=[]
for aTuple in nytWordCount:
    nyttopWords.append(str(aTuple[0]))

wsjtopWords=[]
for aTuple in wsjWordCount:
    wsjtopWords.append(str(aTuple[0]))

foxtopWords=[]
for aTuple in foxWordCount:
    foxtopWords.append(str(aTuple[0]))
```

## Union and Diffs

```
#Checking top words for all networks
union = list(set(cnntopWords) | set(nyttopWords) | set(wsjtopWords) | set(foxtopWords)) 
print("Total Union = ",len(union))
commonTopWords = list(set(cnntopWords) & set(nyttopWords) & set(wsjtopWords) & set(foxtopWords)) 
print("Common in all networks= ", len(commonTopWords))
leftUnion = list(set(cnntopWords) | set(nyttopWords) )
print("Union in Left= ", len(leftUnion))
rightUnion = list( set(wsjtopWords) | set(foxtopWords)) 
print("Union in  Right= ", len(rightUnion))
```

```
cnnnyt = list(set(cnntopWords) & set(nyttopWords) )
print("Common in Left= ", len(cnnnyt))
wsjfox = list( set(wsjtopWords) & set(foxtopWords)) 
print("Common in all Right= ", len(wsjfox))
```

```
#Cross union below - CNN vs Right
cnnuwsj = list(set(cnntopWords) | set(wsjtopWords) )
print("CNN U WSJ= ", len(cnnuwsj))

cnnandwsj = list(set(cnntopWords) & set(wsjtopWords) )
print("CNN AND WSJ= ", len(cnnandwsj))

cnnufox = list( set(cnntopWords) | set(foxtopWords)) 
print("CNN U fOX= ", len(cnnufox))

cnnandfox = list( set(cnntopWords) & set(foxtopWords)) 
print("CNN AND fOX= ", len(cnnandfox))
```

```
#Cross union below - NYTimes vs Right
nytwsj = list(set(nyttopWords) | set(wsjtopWords) )
print("NYT U WSJ= ", len(nytwsj))
nytfox = list( set(nyttopWords) | set(foxtopWords)) 
print("NYT AND WSJ= ", len(nytfox))
```

```
# Checking unique words for each networks
# CNN vs FOX
cnnDiffFox = list( set(cnntopWords) - set(foxtopWords)) 
print("CNN - fOX= ", len(cnnDiffFox))
cnnDiffFox
```
